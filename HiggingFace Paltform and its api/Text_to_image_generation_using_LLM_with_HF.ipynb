```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
```

```python
!pip show torch
```

```python
model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"
model_id2 = "stabilityai/stable-diffusion-xl-base-1.0"

pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16, use_safetensors=True)
pipe = pipe.to("cuda")
```

```python
prompt1= """dreamlikeart, make a beautiful indian lady with proper white skin and is flaunting her curves in a properly indian saree. """
```

```python
image = pipe(prompt1).images[0]
```

```python
print("[PROMPT]:", prompt1)
plt.imshow(image)
plt.axis('off')
```

```python
prompt2 = """ A beautiful beautifull girl is sitting on a chair & She is accompanied by her tiger . make sure to keep them in one frame. """

image = pipe(prompt2).images[0]
```

```python
print("[PROPMT]:", prompt2)
plt.imshow(image)
plt.axis('off')
```

# Working with Stable Diffusion  parameters

1.   Negative Prompting
2.   num_inference_steps
3.   height, weight
4.   num_image_per_prompt



```python
def generate_image(pipe, prompt, params):
  img = pipe(prompt, **params).images

  num_images = len(img)
  if num_images >1:
    fig, ax = plt.subplots(nrows=1, ncols= num_images)
    for i in range(num_images):
      ax[i].imshow(img[i])
      ax[i].axis('off')
  else:
    fig = plt.figure()
    plt.imshow(img[0])
    plt.axis('off')
  plt.tight_layout()

```

```python
prompt = " dreamlike, beautiful girl playing the festivals of colors , drapped in traditonla indian attire, throwing colors on others people in the street. "
params = {}
```

```python
generate_image(pipe, prompt, params)
```

```python
#num_inference_steps
params = {'num_inference_steps': 100}
generate_image(pipe, prompt, params)
```

```python
#height width
params = {'num_inference_steps': 100,'height': 512, 'width': 512}
generate_image(pipe, prompt, params)
```

```python
# num_images_per_prompt
params = {'num_inference_steps': 100,'height': 512, 'width': 512, 'num_images_per_prompt': 4}
generate_image(pipe, prompt, params)
```

```python
#negative prompiting

params = {'num_inference_steps': 100,'height': 512, 'width': 512, 'num_images_per_prompt': 4, 'negative_prompt': 'ugly, distorted, low quality'}
generate_image(pipe, prompt, params)
```
